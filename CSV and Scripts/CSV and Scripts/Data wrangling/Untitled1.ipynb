```python
resultdt = pd.read_csv('AddedRentdata.csv')
display(resultdt.head())
```

```python
postsuburbdata = pd.read_excel('australian_postcodes.xls')
```

```python
display(postsuburbdata.head())
```

```python
subdata = postsuburbdata[["locality","long","lat"]]
```

```python
display(subdata.head())
```

```python
subdata = subdata.rename(columns={'locality':'Suburb','long':'long','lat':'lat'})
```

```python
display(subdata.head())
```

```python
resultdt['Suburb']=resultdt['Suburb'].str.upper()
display(resultdt.head())
finaldata = pd.merge(resultdt,subdata, how = 'inner', on = 'Suburb')
```

```python
display(finaldata)
```

```python
finaldata.to_csv('Results/5147final.csv')
```

```python
averagerent = 471.1448
averagedistance = 31.78897
averageschool = 3.6919
averagecrime = 930.731
averagebuy = 9785736
averagehospital = 1.064368
```

```python
import math
finaldata["Livability rating"] = 0
#display(finaldata)
for i in range(0, len(finaldata)):
    finaldata.iloc[i,9] = math.floor((averagerent/(finaldata.iloc[i,1]+100) 
                                      + averagedistance/(finaldata.iloc[i,2]+30) 
                                      + (finaldata.iloc[i,3]+10)/(averageschool*5) 
                                      + averagecrime/(finaldata.iloc[i,4]+averagecrime)
                                      + averagebuy/(finaldata.iloc[i,5]+averagebuy)
                                      + (finaldata.iloc[i,6]+5)/(averagehospital*3))*100)/100+2
```

```python
display(finaldata)
finadatamodify = finaldata.drop_duplicates('Suburb')
display(finadatamodify)
```

```python
finadatamodify.to_csv('Results/5147finalwithrate.csv')
```
