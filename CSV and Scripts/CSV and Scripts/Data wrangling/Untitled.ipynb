```python
import csv
import pandas as pd
import numpy as np
from IPython.display import display, HTML
currentdata = pd.read_csv('dataforitertion1.csv',index_col=0,header=0)
```

```python
display(currentdata.head())
```

```python
distancedata = pd.read_csv('Melbourne Suburbs by Distance & Direction from CBD - Sheet1.csv')
```

```python
display(distancedata.head())
```

```python
suburbdistance = distancedata[["Suburb","Distance (km)"]]
suburbdistance['Suburb'] = suburbdistance['Suburb'].str.upper()
display(suburbdistance.head())
```

```python
mergedata = pd.merge(suburbdistance,currentdata)
display(mergedata)
```

```python
mergedata.to_csv('iteration1dataupdate1.csv')
```
