```python
import csv
import pandas as pd
import numpy as np
result = pd.read_csv('Results/finalchangeschool&hospital.csv')
#display(result)
```

```python
population = pd.read_csv('EthinicityV4 (1).csv')
```

```python
display(population.head())
```

```python
startnew = [{'Suburb':0,'Chinese_population':0}]
temp = pd.DataFrame(startnew)

for i in range(0, len(population)):
    temp.loc[temp.shape[0]] = {'Suburb':population.loc[i]['Suburb'].replace(' (Vic.)',''),'Chinese_population':int(population.loc[i]['Population'])}
    
display(temp)    
```

```python
finalcheck = pd.merge(temp,result, on='Suburb',how='right')
display(finalcheck)
```

```python
finalcheck.to_csv('Results/finaladdpopulation.csv')
```
