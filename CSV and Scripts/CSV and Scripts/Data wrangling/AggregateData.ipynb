```python
# Crime
#print(crimedata.columns)
crimedata2018 = crimedata.loc[crimedata['Year ending December']==2018]
display(crimedata2018.head())
```

```python
crimeNumSuburb2018 = crimedata2018[["Suburb/Town Name","Incidents Recorded"]]
#display(crimeNumSuburb2018)
crimeN18 = crimeNumSuburb2018.groupby("Suburb/Town Name")["Incidents Recorded"].sum()
suburbc = []
cn = []
for i in crimeN18.index:
    suburbc.append(i)
for i in crimeN18:
    cn.append(i)

crimeN18new = pd.DataFrame({'Suburb':suburbc,'Crime Number':cn})
display(crimeN18new.head())
```

```python
# School

schooldata = pd.read_csv('dv259-allschoolslist-2018.csv')
# print(schooldata.columns)
display(schooldata.head())
```

```python
schoolNumSuburb =  schooldata[["Postal_Town","School_Type"]]
#display(schoolNumSuburb)
schooltype = schoolNumSuburb
schoolPrimary = schoolNumSuburb.loc[schoolNumSuburb['School_Type']=="Primary"]
schoolPS = schoolNumSuburb.loc[schoolNumSuburb['School_Type']=="Pri/Sec"]
schoolSecondary = schoolNumSuburb.loc[schoolNumSuburb['School_Type']=="Secondary"]
schoolSpecial = schoolNumSuburb.loc[schoolNumSuburb['School_Type']=="Special"]
#display(schoolPrimary)

schoolPN = schoolPrimary.groupby('Postal_Town')["School_Type"].count()
schoolPSN = schoolPS.groupby('Postal_Town')["School_Type"].count()
schoolSN = schoolSecondary.groupby('Postal_Town')["School_Type"].count()
schoolSpecialN = schoolSpecial.groupby('Postal_Town')["School_Type"].count()

#display(schoolPSN)
#display(schoolPN.columns)

# create school primiry sum
suburbsp = []
spn = []
for i in schoolPN.index:
    suburbsp.append(i)
#    for j in schoolPSN.index:
#        if i == j:
#            schoolPN[i] = schoolPN[i] + schoolPSN[j]

for i in schoolPN:
    spn.append(i)
    
schoolPNnew = pd.DataFrame({'Suburb':suburbsp,'Primary_School_Number':spn})
# create school secondary sum
suburbss = []
ssn = []
for i in schoolSN.index:
    suburbss.append(i)
#    for j in schoolPSN.index:
#        if i == j:
#            schoolSN[i] = schoolSN[i] + schoolPSN[j]

for i in schoolSN:
    ssn.append(i)
    
schoolSNnew = pd.DataFrame({'Suburb':suburbss,'Secondary_School_Number':ssn})

# create school specail sum
suburbsspecial = []
sspecialn = []
for i in schoolSpecialN.index:
    suburbsspecial.append(i)

for i in schoolSpecialN:
    sspecialn.append(i)
    
schoolSpecialNnew = pd.DataFrame({'Suburb':suburbsspecial,'Special_School_Number':sspecialn})

# create school primary/secondary sum
suburbsps = []
sps = []
for i in schoolPSN.index:
    suburbsps.append(i)

for i in schoolPSN:
    sps.append(i)
    
schoolPSNnew = pd.DataFrame({'Suburb':suburbsps,'PS_School_Number':sps})
#display(schoolPSNnew)

#display(schoolPNnew)
schoolfinal = pd.merge(schoolPNnew, schoolSNnew, how='outer', on=['Suburb'])
#display(schoolfinal)
schoolfinal = pd.merge(schoolfinal, schoolSpecialNnew, how='outer', on=['Suburb'])
#display(schoolfinal)
schoolfinal = pd.merge(schoolfinal, schoolPSNnew, how = 'outer', on = ['Suburb'])
schoolfinal = schoolfinal.fillna(0)
#display(schoolfinal)

schoolfinal['Primary_School_Number']= schoolfinal['Primary_School_Number']+ schoolfinal['PS_School_Number']
schoolfinal['Secondary_School_Number'] = schoolfinal['Secondary_School_Number'] +  schoolfinal['PS_School_Number']
schoolfinal['School_Total'] = schoolfinal['Primary_School_Number']+schoolfinal['Secondary_School_Number']+schoolfinal['Special_School_Number']-schoolfinal['PS_School_Number']
display(schoolfinal)
#CLAYTON = schoolPrimary.loc[schoolPrimary['Postal_Town']=="CLAYTON"]
#display(schoolSNnew)
#display(schoolSNnew)
#display(schoolSpecialNnew)
schoolmerge = schoolfinal[["Suburb","School_Total"]]
display(schoolmerge.head())
```
