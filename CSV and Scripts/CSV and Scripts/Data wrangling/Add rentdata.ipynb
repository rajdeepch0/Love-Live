```python
import re
start = [{'Suburb':0,'Rent2bedroomflat':0}]
temp = pd.DataFrame(start)

for i in range(0, len(rent2bf)):
    
    suburbsplit = re.split(u'-',rent2bf.iloc[i]['Suburb'])
    #print(suburbsplit)
    for e in suburbsplit:
        temp.loc[temp.shape[0]] = {'Suburb':e,'Rent2bedroomflat':int(rent2bf.iloc[i]['Average_rent'].replace('$','').replace('-','0'))}

        
display(temp.head())

start1 = [{'Suburb':0,'Rent3bedroomflat':0}]
temp3bf = pd.DataFrame(start1)

for i in range(0, len(rent3bf)):
    
    suburbsplit = re.split(u'-',rent3bf.iloc[i]['Suburb'])
    #print(suburbsplit)
    for e in suburbsplit:
        temp3bf.loc[temp3bf.shape[0]] = {'Suburb':e,'Rent3bedroomflat':int(rent3bf.iloc[i]['Average_rent'].replace('$','').replace('-','0'))}

        
display(temp3bf.head())

start2 = [{'Suburb':0,'Rent2bedroomhouse':0}]
temp2bh = pd.DataFrame(start2)

for i in range(0, len(rent2bh)):
    
    suburbsplit = re.split(u'-',rent2bh.iloc[i]['Suburb'])
    #print(suburbsplit)
    for e in suburbsplit:
        temp2bh.loc[temp2bh.shape[0]] = {'Suburb':e,'Rent2bedroomhouse':int(rent2bh.iloc[i]['Average_rent'].replace('$','').replace('-','0'))}

        
display(temp2bh.head())

start3 = [{'Suburb':0,'Rent3bedroomhouse':0}]
temp3bh = pd.DataFrame(start3)

for i in range(0, len(rent3bh)):
    
    suburbsplit = re.split(u'-',rent3bh.iloc[i]['Suburb'])
    #print(suburbsplit)
    for e in suburbsplit:
        temp3bh.loc[temp3bh.shape[0]] = {'Suburb':e,'Rent3bedroomhouse':int(rent3bh.iloc[i]['Average_rent'].replace('$','').replace('-','0').replace(',',''))}

        
display(temp3bh.head())
```

```python
rent4bh =  pd.read_csv('suburbrent/suburbrent4bh.csv')

start4 = [{'Suburb':0,'Rent4bedroomhouse':0}]
temp4bh = pd.DataFrame(start4)

for i in range(0, len(rent4bh)):
    
    suburbsplit = re.split(u'-',rent4bh.iloc[i]['Suburb'])
    #print(suburbsplit)
    for e in suburbsplit:
        temp4bh.loc[temp4bh.shape[0]] = {'Suburb':e,'Rent4bedroomhouse':int(rent4bh.iloc[i]['Average_rent'].replace('$','').replace('-','0').replace(',',''))}

        
display(temp4bh.head())
```

```python
rentall =  pd.read_csv('suburbrent/suburbrentall.csv')

start5 = [{'Suburb':0,'RentAverageAll':0}]
tempall = pd.DataFrame(start5)

for i in range(0, len(rentall)):
    
    suburbsplit = re.split(u'-',rentall.iloc[i]['Suburb'])
    #print(suburbsplit)
    for e in suburbsplit:
        tempall.loc[tempall.shape[0]] = {'Suburb':e,'RentAverageAll':int(rentall.iloc[i]['Average_rent'].replace('$','').replace('-','0').replace(',',''))}

        
display(tempall.head())
```

```python
data2change = pd.read_csv('FinalData - After Normalisation,Merging.....csv')
display(data2change.head())
```

```python
finaldata = pd.merge(data2change,temp, how = 'inner', on = 'Suburb')
finaldata = pd.merge(finaldata,temp3bf, how ='inner', on = 'Suburb')
finaldata = pd.merge(finaldata,temp2bh, how ='inner', on = 'Suburb')
finaldata = pd.merge(finaldata,temp3bh, how ='inner', on = 'Suburb')
finaldata = pd.merge(finaldata,temp4bh, how ='inner', on = 'Suburb')
finaldata = pd.merge(finaldata,tempall, how ='inner', on = 'Suburb')

display(finaldata)
```

```python
suburblist = finaldata["Suburb"]
#for e in suburblist:
#    print(e)

schooldata = pd.read_csv('dv259-allschoolslist-2018.csv')
display(schooldata.head())
```

```python
start5 = [{'School_Name':0,'Suburb':0,'School_Type':0,'Education_Sector':0,'Address':0,'Phone_number':0}]
schooltemp = pd.DataFrame(start5)

k = 0
for i in range(0, len(schooldata)):
    for e in suburblist:
        if schooldata.iloc[i]['Address_Town'] == e:
            k = k+1
            schooltemp.loc[schooltemp.shape[0]]={'School_Name':schooldata.iloc[i]['School_Name'],
                                                'Suburb':e,'School_Type':schooldata.iloc[i]['School_Type'],
                                                'Education_Sector':schooldata.iloc[i]['Education_Sector'],
                                                 'Address':str(schooldata.iloc[i]['Address_Line_1'])+', '+str(schooldata.iloc[i]['Address_Town'])
                                                +' '+ str(schooldata.iloc[i]['Address_State'])+' '+str(schooldata.iloc[i]['Address_Postcode']),
                                                 'Phone_number': schooldata.iloc[i]['Full_Phone_No']}
        

print(k)    
display(schooltemp)

```
